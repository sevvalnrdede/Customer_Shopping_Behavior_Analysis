```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

**creating a column**

```python
#age_group
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#purchase_frequency_days
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# Install required libraries

from sqlalchemy import create_engine
from urllib.parse import quote_plus

# SQL Server connection
server = r'.\SQLEXPRESS' # Sunucu Adı kısmındaki değer
database = 'customer_behavior' # Bağlanmak istediğin veritabanı adı

# Note: requires Microsoft ODBC Driver installed separately on your machine
params = quote_plus(
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    f"Trusted_Connection=yes;"
)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

# Write DataFrame to SQL Server
# Read back sample (SQL Server uses TOP instead of LIMIT)
try:
    df_check = pd.read_sql(f"SELECT TOP 5 * FROM customer", engine)
    print(df_check)
except Exception as e:
    print("Bağlantı hatası:", e)
```

```python
df.columns = [str(col).strip() if col is not None else f"column_{i}" for i, col in enumerate(df.columns)]
df.columns = [col if col != "" else f"col_{i}" for i, col in enumerate(df.columns)]
print(df.columns)
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.to_sql("customer", engine, if_exists="replace", index=False)
print("Table is in SQL")
```
